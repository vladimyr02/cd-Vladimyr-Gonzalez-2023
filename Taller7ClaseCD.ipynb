```python
x = df.iloc[:,[2,3,4,5,6,7,8,9,10]]
y = df.Survived
logit_model=sm.Logit(y,x)
result=logit_model.fit()
print(result.summary())
```

```python
import statsmodels.api as sm
import itertools
import pandas as pd

# Definir las variables independientes
X_columns = [2, 3, 4, 5, 6, 7, 8, 9, 10]

# Crear una lista de diccionarios para registrar las combinaciones y sus métricas
results = []

# Iterar a través de todas las combinaciones posibles de variables independientes
for r in range(1, len(X_columns) + 1):
    combinations = itertools.combinations(X_columns, r)

    for combo in combinations:
        X_subset = df.iloc[:, list(combo)]
        logit_model = sm.Logit(y, X_subset)
        result = logit_model.fit()

        # Calcular el R-squared como métrica
        r_squared = 1.0 - result.llf / result.llnull

        # Registrar la combinación y su métrica en la lista
        results.append({'Combination': combo, 'R-squared': r_squared})

# Convertir la lista de resultados en un DataFrame
results_df = pd.DataFrame(results)

# Encontrar la mejor combinación basada en el R-squared
best_combination = results_df.loc[results_df['R-squared'].idxmax()]

# Mostrar la mejor combinación y su métrica
print("Mejor combinación de variables:", best_combination['Combination'])
print("Mejor R-squared:", best_combination['R-squared'])
```

```python
import statsmodels.api as sm

# Selecciona las columnas de características en el conjunto de entrenamiento
x_train = train_df.iloc[:, [2, 3, 4, 5, 6, 7, 8, 9, 10]]

# El objetivo 'y_train' debe ser de train_df, no 'y'
y_train = train_df['Survived']

# Ajusta el modelo de regresión logística
logit_model = sm.Logit(y_train, sm.add_constant(x_train))  # Agrega una columna constante para el término constante
result = logit_model.fit()
print(result.summary())
```

```python
# Realiza predicciones en el conjunto de prueba (o cualquier otro conjunto de datos)
x_train = train_df.iloc[:, [2, 3, 4, 5, 6, 7, 8, 9, 10]]
y_pred_prob = result.predict(sm.add_constant(x_train))

# Convierte las probabilidades en etiquetas binarias (0 o 1) usando un umbral
threshold = 0.5
y_pred_binary = (y_pred_prob > threshold).astype(int)

# Imprime las predicciones
print(y_pred_binary)
```

```python
# Inicializa los contadores
TP_count = 0
FP_count = 0
FN_count = 0

# Define el umbral para convertir las probabilidades en etiquetas binarias
threshold = 0.5

# Itera a través de los valores reales y las predicciones
for y_real, y_prob in zip(y_train, y_pred_binary):
    # Convierte la probabilidad en etiqueta binaria
    y_pred_binary = 1 if y_prob > threshold else 0

    # Compara con el valor real y actualiza los contadores
    if y_real == 1 and y_pred_binary == 1:
        TP_count += 1
    elif y_real == 0 and y_pred_binary == 1:
        FP_count += 1
    elif y_real == 1 and y_pred_binary == 0:
        FN_count += 1

# Imprime los resultados
print(f'Verdaderos Positivos (TP): {TP_count}')
print(f'Falsos Positivos (FP): {FP_count}')
print(f'Falsos Negativos (FN): {FN_count}')
```

```python
# Calcula la precisión y la exhaustividad
precision = TP_count / (TP_count + FP_count)
recall = TP_count / (TP_count + FN_count)

# Imprime los resultados
print(f'Precisión (Precision): {precision:.2f}')
print(f'Exhaustividad (Recall): {recall:.2f}')
```
