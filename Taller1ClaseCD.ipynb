```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Datos
area = [76, 92, 107, 42, 65, 150, 65, 105, 40, 55]
valor = [200, 320, 300, 160, 190, 315, 230, 380, 155, 175]

# Convertir las listas a arreglos de NumPy
area_np = np.array(area).reshape(-1, 1)
valor_np = np.array(valor)

# Crear y ajustar el modelo de regresión lineal
regression_model = LinearRegression()
regression_model.fit(area_np, valor_np)

# Obtener coeficientes
pendiente = regression_model.coef_[0]
intercepto = regression_model.intercept_

# Imprimir los coeficientes
print(f"Pendiente: {pendiente}")
print(f"Intercepto: {intercepto}")

# Predecir valores basados en el modelo
valores_predichos = regression_model.predict(area_np)

# Graficar los datos y la regresión lineal
plt.scatter(area, valor, label='Datos')
plt.plot(area, valores_predichos, color='red', label='Regresión lineal')
plt.xlabel('Área')
plt.ylabel('Valor')
plt.legend()
plt.show()
```
