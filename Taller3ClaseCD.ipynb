```python
df0 = df[df["BuildingArea"]<=800]
df0
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
df0.plot(x = 'BuildingArea', y = 'Price', c = 'navy', kind = "scatter", ax = ax)
ax.set_title('Distribución de precio y area construida');
```

```python
dfzero = df0[df0["BuildingArea"]<5]
len(dfzero)
```

```python
df0 = df0[df0["BuildingArea"]>=5]
df0
```

```python
df0 = df0[(np.abs(zscore(df0['BuildingArea']))<=3)]
df0
```

```python
import itertools
X = df0.drop(columns=['Price','Lattitude','Longtitude','Postcode','Bedroom2'])
y = df0['Price']

best_r2 = 0
best_features = None

# Iterar sobre todas las combinaciones de tres variables
for combo in itertools.combinations(X.columns, 3):
    X_subset = X[list(combo)]

    model = LinearRegression()
    model.fit(X_subset, y)

    y_pred = model.predict(X_subset)
    r2 = r2_score(y, y_pred)

    if r2 > best_r2:
        best_r2 = r2
        best_features = combo

print("\nLas mejores variables con el mayor R2 son:", best_features)
print("R2 correspondiente:", best_r2)
```
